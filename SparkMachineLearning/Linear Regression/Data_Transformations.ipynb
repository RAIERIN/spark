## Data Features
### StringIndexer
We often convert string information into numerical information as a categorical feature. This is easily done with the StringIndexer Method:

```python
from pyspark.ml.feature import StringIndexer
```

```python
df = spark.createDataFrame(
    [(0, "a"), (1, "b"), (2, "c"), (3, "a"), (4, "a"), (5, "c")],
    ["user_id", "category"])
```

```python
df.show()
```

```python
indexer = StringIndexer(inputCol="category", outputCol="categoryIndex")
indexed = indexer.fit(df).transform(df)
```

```python
indexed.show()
```

### VectorIndexer

VectorAssember is a transformer that combines a given list of columns into a single vector column. It is useful for combining raw features and features generated by different feature transformers into a single feature vector. In order to train ML model like logistic regression and decision tree. VectorAssembeler accepts the following input column types: all numeric types, bool type and vector type. In each row, the values of the input columns will be concatenated into a vector in the specified order.
Assume that we have a DataFrame with the columns id, hour, mobile, userFeatures, and clicked:

 id | hour | mobile | userFeatures     | clicked
----|------|--------|------------------|---------
 0  | 18   | 1.0    | [0.0, 10.0, 0.5] | 1.0
 
 userFeatures is a vector column that contains three user features. We want to combine hour, mobile and userFeatures into a single feature vector called features and use it to predict clicked or not. If we set VectorAssembler's input columns to hour, mobile, and userFeatures and output column to features, after transformation we should get the following DataFrame:
 
  id | hour | mobile | userFeatures     | clicked | features
----|------|--------|------------------|---------|-----------------------------
 0  | 18   | 1.0    | [0.0, 10.0, 0.5] | 1.0     | [18.0, 1.0, 0.0, 10.0, 0.5]

```python
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
```

```python
dataset = spark.createDataFrame(
    [(0,18,1.0, Vectors.dense([0.0,10.0,0.5]), 1.0)],
    ["id","hour","mobile","userFeatures","clicked"]
)
```

```python
dataset.show()
```

```python
assembler = VectorAssembler(
    inputCols=["hour","mobile","userFeatures"],
    outputCol="features"
)

output= assembler.transform(dataset)
print("Assembled columns 'hour','mobile','userFeatures', to vector column 'features'")
output.select('features',"clicked").show()
```
